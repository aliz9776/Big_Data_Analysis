load Y1 and Y2 dataset

```python
import pandas as pd
import numpy as np
# Load dataset Y1
y1_data = pd.read_csv("Y1.csv")
y1 = y1_data.values
print(y1_data)
# Load dataset Y2
y2_data = pd.read_csv("Y2.csv")
y2 = y2_data.values
print(y2_data)
# Calculate transpose of Y1 and Y2
y1 = y1.T
y2 = y2.T
```

calculate covariance of Y1 matrix as C

```python
C=np.cov(y1)
print(C)
```

calculate svd form of matrix c and after calculate matrux M with that for Y1 matrix

```python
from scipy.linalg import sqrtm

#calculate svd form of matrix cov

U_c, S_c, V_c_T = np.linalg.svd(C)
Sigma_c = np.diag(S_c)


print(" matrix U_C : ")

print(U_c)

print(" matrix Sigma_C : ")

print(np.diag(Sigma_c))

print(" matrix V_c_T : ")

print(V_c_T)


#using relation between sigma C and sigma M from part a
Sigma_c_sqrt = sqrtm(Sigma_c)

# matrix M calculation for y1

M = np.dot(np.dot(U_c,Sigma_c_sqrt),V_c_T)
print('matrix M for y1 :')
print(M)

print('matrix M transpose for y1 :')
print(M.T)
```

make x smaples in gausian distribution with mean = [0, 0] and cov = [[1, 0], [0, 1]]

```python
import numpy as np
import matplotlib.pyplot as plt
# Set the parameters for the Gaussian distribution
mean = [0, 0]
covariance = [[1, 0], [0, 1]]

# Number of samples
n = 50000  # You can adjust this value as needed

# Generate samples from the Gaussian distribution
samples = np.random.multivariate_normal(mean, covariance, n)

# Create the matrix X
x = samples.T

print("Matrix X:")
print(x)
print(x.shape)




```

calculate matrix Y1 with M matrix we calculate

```python
# y1 = M . X
y1_predict = np.dot(M , x)
```

plot orginal Y1 AND Y1 that we predict with matrix M

```python
# Create subplots for y1_convert and y1 side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot y1_convert
ax1.scatter(y1_predict[0], y1_predict[1])
ax1.set_title('y1_predict')
ax1.set_xlabel('y1_predict[x]')
ax1.set_ylabel('y1_predict[y]')

# Plot y1
ax2.scatter(y1[0], y1[1])
ax2.set_title('y1')
ax2.set_xlabel('y1[x]')
ax2.set_ylabel('y1[y]')

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.4)

# Display the plot
plt.show()
```

calculate covariance of Y2 matrix as C

```python
C=np.cov(y2)
print(C)
```

calculate svd form of matrix c and after calculate matrix M with that for Y2 matrix

```python
from scipy.linalg import sqrtm

#calculate svd form of matrix cov

U_c, S_c, V_c_T = np.linalg.svd(C)
Sigma_c = np.diag(S_c)


print(" matrix U_C : ")

print(U_c)

print(" matrix Sigma_C : ")

print(np.diag(Sigma_c))

print(" matrix V_c_T : ")

print(V_c_T)


#using relation between sigma C and sigma M from part a
Sigma_c_sqrt = sqrtm(Sigma_c)

# matrix M calculation for y2

M = np.dot(np.dot(U_c,Sigma_c_sqrt),V_c_T)
print('matrix M for y2 :')
print(M)

print('matrix M transpose for y2 :')
print(M.T)
```

calculate matrix Y2 with transpose of M matrix we calculate

```python

# y2 = MT . X
y2_predict = np.dot(M.T , x)
```

plot orginal Y2 AND Y2 that we predict with matrix M

```python
# Create subplots for y2_convert and y2 side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot y2_convert
ax1.scatter(y2_predict[0], y2_predict[1])
ax1.set_title('y2_predict')
ax1.set_xlabel('y2_predict[x]')
ax1.set_ylabel('y2_predict[y]')

# Plot y2
ax2.scatter(y2[0], y2[1])
ax2.set_title('y2')
ax2.set_xlabel('y2[x]')
ax2.set_ylabel('y2[y]')

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.4)

# Display the plot
plt.show()
```
